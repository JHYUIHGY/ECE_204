**Problem 3.** In this Question you need to find how to limit the y axes to a certain value.

What argument and value can you use in df.plot.scatter() to limit the y-axis so that it only goes up till 10?

```python
df = pd.read_csv("worldHappinessReport_2018.csv")
ax = df.plot.scatter(x="economy", y="happiness", figsize=(7,5), ylim=(0,10), xlim=(4,13))
ax.set_title("Happiness vs Economy Distribution for Countries in Year 2018", fontsize=15)
ax.set_xlabel("Economy(Log GDP Per Capita)", fontsize=13)
ax.set_ylabel("Happiness", fontsize=13)
fig = ax.get_figure()
fig.savefig("worldHappinessReport.png")
```

**Problem 4.** In this question, you need to understand the difference between default and explicit plotting modes in pandas as discussed in class.

```python
df4 = pd.read_csv("nyc_year_fare_proc.csv")
df4.set_index("HourOfDay", inplace=True)
ax = df4.plot.line(legend=True , figsize=(10,5))
ax.set_title("Taxi Fare Distribution in NYC Over the Years", fontsize=18)
ax.set_xlabel("Hour of Day", fontsize=13)
ax.set_ylabel("Taxi Fare (USD)", fontsize=13)
```

```python
# Define the coordinates of the points for each letter
A_x = [-4, -4, -4, -4, -4, -3, -2, -1, -3, -2, -1, -3, -2, -1]
A_y = [4, 3, 2, 1, 0, 4, 4, 4, 2, 2, 2, 0, 0, 0]
B_x = [6, 6, 6, 6, 6, 7, 8, 9, 7, 8, 9, 7, 8, 9]
B_y = [4, 3, 2, 1, 0, 4, 4, 4, 2, 2, 2, 0, 0, 0]
C_x = [1, 1, 1, 1, 1, 2, 3, 4, 2, 3, 4]
C_y = [4, 3, 2, 1, 0, 4, 4, 4, 0, 0, 0]

dfA = pd.DataFrame({'x': A_x, 'y':A_y})
dfB = pd.DataFrame({'x': B_x, 'y':B_y})
dfC = pd.DataFrame({'x': C_x, 'y':C_y})

axA = dfA.plot.scatter(x='x', y='y')
dfB.plot.scatter(x='x', y='y', ax=axA) # ax=axA ensure this plot is made on the same axis as the first plot!
dfC.plot.scatter(x='x', y='y', ax=axA) # ax=axA ensure this plot is made on the same axis as the first plot!
```
