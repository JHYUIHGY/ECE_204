**Problem 3.**
> Read the dataset `movies.csv` and drop missing values if present. We want to find out the average earnings of a movie. For this purpose we will remove the outliers (The very high grossing "blockbuster movies").
>
>**Report the mean of the `US_Gross` column after doing the outlier removal**. Consider any data outside 3 standard deviations of the mean as an outlier. 
>
>`NOTE:` Make sure to round your output to 2 decimal places.

```python
df = pd.read_csv('movies.csv')
df
```

```python
df = df.dropna()
df
```

```python
m = df.US_Gross.mean()
s = df.US_Gross.std()
df = df[(df.US_Gross > m-3*s) & (df.US_Gross < m+3*s)]
mean_gross = df.US_Gross.mean().item() 
print(round(mean_gross, 2))
mean_gross
```
