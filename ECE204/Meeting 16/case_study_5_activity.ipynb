**Problem 1.** Read the `blobs_train.csv` and `test_data.csv` dataset. Fit a DecisionTreeClassifier (`random_state = 1, max_depth=2`)  model on the training data. Predict on the testing data and plot the results. **What rules does the  decision tree learn for `feat1` and `feat2`? Choose the correct answer from the options**

Please use the `cmap="viridis"` as we have above.

```python
blobs_train = pd.read_csv("blobs_train.csv")
test_data = pd.read_csv("test_data.csv")

blobs_train.head()
```

```python
# your code here

clf = DecisionTreeClassifier(random_state=1, max_depth=2)
clf.fit(blobs_train[["feat1", "feat2"]], blobs_train["label"])

# predict the label on test data
test_data["pred"] = clf.predict(test_data[["feat1", "feat2"]])

# plot both of the data in one plot
# We are making the test point 10% transparent using alpha = 0.1
# Dark points are training points
ax1 = test_data.plot.scatter(x="feat1",y="feat2",c="pred", cmap="viridis", colorbar=False, figsize=(8,6), alpha=0.1)
blobs_train.plot.scatter(x="feat1",y="feat2",c="label", cmap="viridis", colorbar=False, figsize=(8,6), ax=ax1)
```
