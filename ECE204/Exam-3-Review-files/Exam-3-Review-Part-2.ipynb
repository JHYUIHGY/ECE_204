```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import pandas as pd

df = pd.read_csv('blobs_2.csv')
df.plot.scatter(x="foo", y="bar")
```

```python
# Your code here
km = KMeans(n_clusters=2, random_state=42)
X = km.fit(df)
y_hat = X.predict(df)

print("No. of points in cluster 1:", sum(y_hat))
print("No. of points in cluster 2:", len(y_hat) - sum(y_hat))
centre1 = X.cluster_centers_[0]
centre2 = X.cluster_centers_[1]

import numpy as np

new_point = np.array([2, 2])
print(np.linalg.norm(centre1 - new_point))
print(np.linalg.norm(centre2 - new_point))


```
