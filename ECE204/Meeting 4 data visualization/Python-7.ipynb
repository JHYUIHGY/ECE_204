```python
df = pd.read_csv("google_search_interest.csv")
ax = df.plot.line(x="month", y=["data science","python"], figsize=(10,5))
ax.set_title("Search Interest from Feb 2010 to Feb 2021", fontsize=18)
ax.set_xlabel("Time(months)", fontsize=13)
ax.set_ylabel("Search Interes(data science/python)", fontsize=13)
ax.legend(["Data Science", "Python"])
fig = ax.get_figure()
fig.savefig("google_search_interest.png")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the coordinates of the points for each letter
A_x = np.array([-4, -4, -4, -4, -4, -3, -2, -1, -3, -2, -1, -3, -2, -1])
A_y = np.array([4, 3, 2, 1, 0, 4, 4, 4, 2, 2, 2, 0, 0, 0])
B_x = np.array([4, 4, 4, 4, 4, 5, 6, 7, 5, 6, 7, 5, 6, 7])
B_y = np.array([4, 3, 2, 1, 0, 4, 4, 4, 2, 2, 2, 0, 0, 0])
C_x = np.array([0, 0, 0, 0, 0, 1, 2, 3, 1, 2, 3])
C_y = np.array([4, 3, 2, 1, 0, 4, 4, 4, 0, 0, 0])


# Plot the points using scatter
plt.scatter(A_x, A_y, s=100, c='r', marker='o')
plt.scatter(B_x, B_y, s=100, c='b', marker='o')
plt.scatter(C_x, C_y, s=100, c='g', marker='o')


# Set the axes limits and labels
plt.xlim(-5, 8)
plt.ylim(-1, 5)
plt.xlabel('x')
plt.ylabel('y')

# Show the plot
plt.show()
```
